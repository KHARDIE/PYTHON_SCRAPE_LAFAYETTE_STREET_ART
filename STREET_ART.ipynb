```python
import csv
import re
import time
import requests
from bs4 import BeautifulSoup
import urllib2
import urllib
#import socket
from HTMLParser import HTMLParser

fieldlist =[]
  
fieldlist.insert(0,('NID', 'CITY_WEBSITE', 'NAME','ARTIST','ARTIST WEBSITE','ARTIST_LOCATION', 'STREETART_LOCATION','MATERIALS', 'PRICE', 'STATEMENT'))

with open('STREET_ART_ID.csv', 'r') as f:
    for line in f:
        PTiD = line
#         print PTiD
        
#         url_values = urllib.urlencode(data)
        url = "http://www.cityoflafayette.com/index.aspx?NID=" + str(PTiD)
        res = requests.get(url, stream = True)
        raw_html = res.content
        
        user_agent = 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)'
        headers = { 'User-Agent' : user_agent }
        req = urllib2.Request(url)
        
            ## With error checking
        try:
            response = urllib2.urlopen(req)
            siteurl = response.geturl()
            siteinfo = response.info()
            #print(siteinfo)
            #print(siteurl)
        except urllib2.HTTPError as e:
            print 'The server couldn\'t fulfill the request.'
            print 'Error code: ', e.code
        except urllib2.URLError as e:
            print 'We failed to reach a server.'
            print 'Reason: ', e.reason
        else:
            # everything is fine
            the_page = response.read()     

        soup = BeautifulSoup(the_page)
#         print (soup.prettify)

        NID = PTiD
        NID = NID.encode('ascii','ignore')
        NID = NID.strip()      
        print "NID: "+ str(NID)

        CITY_WEBSITE = url.encode('ascii','ignore')
        CITY_WEBSITE = CITY_WEBSITE.strip()      
        print "CITY WEBSITE: "  + CITY_WEBSITE
    
        NAME = soup.find('title')
        NAME = NAME.text[38:-1]
        NAME = NAME.split('by ')[0]
        NAME = NAME.encode('ascii','ignore')
        NAME = NAME.strip()
        print "NAME: " + NAME
        
        ARTIST = soup.find('title')
        ARTIST = ARTIST.text[38:-1]
        ARTIST = ARTIST.split('by ')[1]
        ARTIST = ARTIST.encode('ascii','ignore')
        ARTIST = ARTIST.strip()
        print "ARTIST: " +ARTIST

        ARTIST_WEBSITE = soup.find('a',href=True, attrs={'class': 'Hyperlink'})
        ARTIST_WEBSITE = str(ARTIST_WEBSITE)
        ARTIST_WEBSITE = ARTIST_WEBSITE.replace('<a class="Hyperlink" href="','')
        ARTIST_WEBSITE = ARTIST_WEBSITE.split('" target="')[0]
        ARTIST_WEBSITE = ARTIST_WEBSITE.replace('http://www.cityoflafayette.com/index.aspx?nid=795','NONE')
        print "ARTIST WEBSITE: " + ARTIST_WEBSITE

# Look in TD for elements needed

        tds = soup.find_all('td')
        for ind, itm in enumerate(tds):
            if ind == 8:
                onlytext = itm.text
#                 print ind, onlytext
       
#             Initialize the Indexes
                ArtistIndex = onlytext.index('Artist')
                StreetArtLocationIndex = onlytext.index('Location:')
                MaterialsIndex = onlytext.index('Materials:')
                PriceIndex = onlytext.index('Price:')
                StatementIndex = onlytext.index('Artists Statement')if 'Artists Statement' in onlytext else 'NONE'
                StatementIndex2 = onlytext.index('Artist Statement:')if 'Artist Statement:' in onlytext else 'NONE'
                EndIndex = onlytext.index('View the video')if 'View the video' in onlytext else 'NONE'
                
                ARTIST = onlytext[ArtistIndex +7:StreetArtLocationIndex]
                ARTIST = ARTIST.encode('ascii','ignore')
                ARTIST = ARTIST.strip()
                if ('\n' in ARTIST):
                    ARTIST_LOCATION = ARTIST.split('\n')[1]
                    ARTIST_LOCATION = ARTIST_LOCATION.strip()
                else:
                    ARTIST_LOCATION = 'NONE'
                print "ARTIST LOCATION: " + ARTIST_LOCATION
                
                STREETART_LOCATION = onlytext[StreetArtLocationIndex+9:MaterialsIndex ]
                STREETART_LOCATION = STREETART_LOCATION.encode('ascii','ignore')
                STREETART_LOCATION = STREETART_LOCATION.strip()
                STREETART_LOCATION = STREETART_LOCATION.replace("\n",'')
                STREETART_LOCATION = STREETART_LOCATION.split('\r')[0]
                print "STREETART LOCATION: "+ STREETART_LOCATION
                
                MATERIALS = onlytext[MaterialsIndex+10:PriceIndex ]
                MATERIALS = MATERIALS.encode('ascii','ignore')
                MATERIALS = MATERIALS.strip()
                MATERIALS = MATERIALS.replace("\n",'')
                MATERIALS = MATERIALS.replace("\r",'')
                MATERIALS = MATERIALS.split('Size:')[0]
                print "MATERIALS: "+ MATERIALS
                
                PRICE = onlytext[PriceIndex+6: StatementIndex-8 if 'Artists Statement' in onlytext else PriceIndex+15:]
                PRICE = PRICE.encode('ascii','ignore')
                PRICE = PRICE.strip()
                PRICE = PRICE.replace("\n",'')
                print "PRICE: "+ PRICE               
                
                if StatementIndex == 'NONE':
                    STATEMENT = onlytext[StatementIndex2+17: EndIndex]
                    STATEMENT = STATEMENT.encode('ascii','ignore')
                    STATEMENT = STATEMENT.strip()
                    STATEMENT = STATEMENT.replace(":",'')
                    STATEMENT = STATEMENT.replace('"','')                    
                    STATEMENT = STATEMENT.replace("\n",'')
                    STATEMENT = STATEMENT.replace("\r",'')
                    STATEMENT = STATEMENT.replace('             ','')
                    STATEMENT = STATEMENT.strip()
                else:
                    STATEMENT = onlytext[StatementIndex+17: EndIndex if 'View the video' in onlytext else StatementIndex+300:]
                    STATEMENT = STATEMENT.encode('ascii','ignore')
                    STATEMENT = STATEMENT.strip()
                    STATEMENT = STATEMENT.replace(":",'')
                    STATEMENT = STATEMENT.replace('"','')  
                    STATEMENT = STATEMENT.replace("\n",'')
                    STATEMENT = STATEMENT.replace("\r",'')
                    STATEMENT = STATEMENT.replace('             ','')
                    STATEMENT = STATEMENT.strip()
                print "STATEMENT: "+ STATEMENT
                           
        fieldlist.append((NID, CITY_WEBSITE, NAME,ARTIST,ARTIST_WEBSITE, ARTIST_LOCATION, STREETART_LOCATION,MATERIALS, PRICE, STATEMENT))
print fieldlist    

        
#          ###
#         #   #
#           #
#         #  
#         ######        

# # Step Two, Write out a list
     
with open('STREET_ART_OUT.csv', 'wb') as csvfile:
    fieldwriter = csv.writer(csvfile, delimiter=',',quotechar='"', quoting=csv.QUOTE_MINIMAL)
    fieldwriter.writerows(fieldlist)
```
